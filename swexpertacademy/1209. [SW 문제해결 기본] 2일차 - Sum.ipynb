```python
import sys
sys.stdin = open('data/1209.txt', 'r')

for test_case in range(1,11):
    trash = sys.stdin.readline()
    answer = []
    matrix = []
    for i in range(100):
        mat = list(map(int,sys.stdin.readline().split()))
        matrix.append(mat)
    
    b,c = 0,0
    for i in range(100):
        answer.append(sum(matrix[i]))
        a = 0
        for j in range(100):
            a += matrix[j][i]
            if i==j :
                b += matrix[i][j]
            if i+j == 99:
                c += matrix[i][j]
        answer.append(a)
    answer.append(b)
    answer.append(c)
    print(f'#{test_case} {max(answer)}')
```
